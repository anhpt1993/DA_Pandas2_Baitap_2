# 0. Import các thư viện cần thiết

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Đọc dữ liệu từ file CSV

```python
df = pd.read_csv("https://s3.ap-southeast-1.amazonaws.com/learn.codegym.vn/media/public/2021/12/31011319/OnlineRetail.csv", encoding = "ISO-8859-1")
df.head()
```

```python
df.info()
```

# 2. Xây dựng bảng Pivot Table, tính trung bình cộng số lượng các mặt hàng theo mỗi số hoá đơn theo từng quốc gia

```python
df1 = df.pivot_table(values = "Quantity", index = "InvoiceNo", columns = "Country", aggfunc = "mean")
df1
```

# 3. Xây dựng bảng Pivot Table, với mỗi khách hàng cho biết số lượng mua hàng lớn nhất và nhỏ nhất theo kho

```python
df2 = df.pivot_table(values = "Quantity", index = "CustomerID", aggfunc = {min, max})
df2
```

# 4. Xây dựng bảng Pivot Table, với mỗi mã kho tính tổng số lượng các mặt hàng và trung bình cộng giá

```python
df3 = df.pivot_table(values = ["Quantity", "UnitPrice"], index = "StockCode", aggfunc = {"Quantity":np.sum, "UnitPrice":np.mean})
df3.rename(columns = {"Quantity":"Tổng số lượng", "UnitPrice":"Trung bình cộng giá"}, inplace = True)
df3
```

# 6. Xây dựng bảng Pivot Table cho biết tổng số lượng hàng bán được của mỗi ngày

```python
df4 = df.pivot_table(values = "Quantity", index = "InvoiceDate", aggfunc = "sum")
df4.rename(columns = {"Quantity": "Total Quantities"}, inplace = True)
df4
```
